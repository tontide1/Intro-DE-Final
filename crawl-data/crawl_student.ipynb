```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import re
```

```python
url = 'https://fit.iuh.edu.vn/'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
list=[]
if response.status_code == 200:
    ul_tag = soup.select_one('#menu-vn-main-menu')
    
    if ul_tag:
        a_tags = ul_tag.find_all('a')
        print(a_tags)
        for a_tag in a_tags:
            new_url=url+a_tag.get("href")
            list.append(new_url)
            rp=requests.get(new_url)
            sp=BeautifulSoup(rp.content,'lxml')
            if rp.status_code==200:
                data=sp.select_one('div.left-content')
                if data is not None:
                    print(data.text)
                    list.append({
                        '':data.text.strip()
                    })


print(list)     
```
