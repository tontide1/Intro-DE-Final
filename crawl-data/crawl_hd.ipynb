```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import re
import csv
```

```python
url ='https://fit.iuh.edu.vn/news.html@215@Doan-Thanh-nien/'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
daotao=[]
if response.status_code == 200:
    results = soup.select_one('div.content-list')
    data=results.select('div.content')
    if data:
        for i in data:
            new_url='https://fit.iuh.edu.vn/'+i.select_one('a')['href']
            response2=requests.get(new_url)
            soup2=BeautifulSoup(response2.content, 'html.parser')
            if response2.status_code==200:
                data=soup2.select_one('div.content-list')
                print(data.text.strip())
                daotao.append(data.text.strip())
df=pd.DataFrame(daotao)
df.to_csv('hoatdong.csv', index=False, header=False)
            
```
