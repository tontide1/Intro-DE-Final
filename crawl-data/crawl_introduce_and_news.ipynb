```python
from bs4 import BeautifulSoup
import requests
import pprint 
import csv
import pandas as pd
import string
```

```python
url = 'https://fit.iuh.edu.vn/'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
data_list = []
if response.status_code == 200:
    ul_tag = soup.select_one('#menu-vn-main-menu')
    if ul_tag:
        a_tags = ul_tag.find_all('a')
        print(a_tags)
        for a_tag in a_tags:
            new_url = url + a_tag.get("href")
            rp = requests.get(new_url)
            sp = BeautifulSoup(rp.content, 'lxml')
            if rp.status_code == 200:
                data = sp.select_one('div.left-content')
                if data is not None:
                    content = data.text.strip()
                    content = content.replace('\t', ' ')
                    content = content.replace('\n', '')
                    data_list.append({'': content})

df = pd.DataFrame(data_list)
df.to_csv('introduce.csv', index=False)
```
